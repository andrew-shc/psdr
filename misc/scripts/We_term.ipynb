```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
H, W = 1080, 1920
n_cam = np.array([0,0,1])  # shape (3,)
ys = np.linspace(-1, 1, H)*(H/W)
xs = np.linspace(-1, 1, W)
xv, yv = np.meshgrid(xs, ys)
focal_length = 0.5
positions = np.stack([xv, yv, np.ones_like(xv)*focal_length], axis=-1)
omega = positions / np.linalg.norm(positions, axis=-1, keepdims=True)

# Dot product along the last dimension
cos_term = np.dot(omega, n_cam)  # shape (H, W)

plt.imshow(cos_term, cmap='gray')
```
